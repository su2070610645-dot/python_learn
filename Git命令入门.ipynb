```python
# 推送到远程github仓库
git push origin master   # 推送到 master 分支
# 或者
git push origin main     # 推送到 main 分支
# master分支和main分支有什么区别？
# 历史背景：master 是传统的默认分支名，但近年来由于"master"这个词可能带有奴隶制的历史含义，许多项目（包括GitHub）开始使用 main 作为默认分支名
# 功能上：两者没有本质区别，都是默认分支，只是名称不同
# 现代趋势：新创建的仓库默认使用 main，但你可以通过设置将默认分支改为 master 或其他名称
# 可以通过查看github上仓库中主分支名称来确认
```

```python
# 查看当前分支
git branch
# 创建新分支，但不切换
git branch 新分支名称
# 创建并切换到新分支
git checkout -b 新分支名
# 更新的命令，创建并切换
git switch -c 新分支名
# 切换到其他分支
git checkout 分支名
# 删除本地分支 注意：你不能删除你当前所在的分支。如果需要删除当前分支，请先切换到其他分支（如 main 或 master）
# 安全删除
git branch -d 分支名
# 强制删除
git branch -D 分支名
# 删除远程分支
git push origin --delete 分支名
# 清理本地无效的远程分支缓存
git fetch -p
```

```python
# 首次推送需要设置 upstream
git push --set-upstream origin main
# 作用：
#    告诉 Git：将本地分支与远程仓库的特定分支关联起来
#    之后可以直接使用 git push（不带参数）推送，Git 会知道推送到哪个远程分支
# 设置后，git pull 也会自动从对应的远程分支拉取
# 为什么需要：
#    首次推送时，Git 不知道你的本地分支应该对应远程仓库的哪个分支，所以需要明确指定
```

```python
# 一个完整的git流程
git remote add origin <repository-url>
git init 
git add .
git commit -m "描述本次修改"
git push origin master
```

```python
# 注意如果仓库是从远程克隆，需要先拉取远程分支
git pull
```

这个警告是因为git检测到文件中使用了LF换行符(Linux/macOS风格)，但我的git配置可能会在下次处理时将其转换为CRLF(Windows风格)。这是跨平台协作时的常见问题。解决方案如下：

1、配置git自动处理换行符

Windows: git config --global core.autocrlf true

Linux/macOS：git config --global core.autocrlf input

跨平台项目：git config --global core.autocrlf true
